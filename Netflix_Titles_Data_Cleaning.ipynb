# Rename column headers to be clean and uniform

```python
df.columns = df.columns.str.strip().str.lower()
df.columns
```

# Standardize text values


```python
df['country'] = df['country'].str.strip().str.title()
df['director'] = df['director'].str.strip().str.title()
df['cast'] = df['cast'].str.strip().str.title()
```

# Convert date formats to a consistent type (e.g., dd-mm-yyyy).
# Check and fix data types

```python
df['date_added'] = pd.to_datetime(df['date_added'].str.strip(), errors='coerce')
```

# Identify and handle missing values using .isnull() in Python

There is more null value in director column so i apply mode function to fill most of it by grouping Type and Country


```python
df['director'] = df['director'].fillna(df.groupby(['type', 'country'])['director'].transform(lambda x: x.mode().dropna().iloc[0] if not x.mode().dropna().empty else None))
```

```python
df.info()
```

 Identify Missing Values

```python
print(df.isnull().sum())
```

Handle Missing Values

```python
df = df.dropna()
```

```python
print(df.isnull().sum())
```

 Remove Duplicate Rows

```python
df = df.drop_duplicates()
```
